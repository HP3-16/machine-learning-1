```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return 3*x**2 - 5*x + 2
```

```python
# func(15.0)

x_ = np.arange(-5,5,0.5)
y_ = func(x_)
plt.plot(x_,y_)
```

Basics of Derivative


$\displaystyle \lim_{x \to \infty} f(x+h) - f(x) / h$

```python
x = 10
h = 0.00000001
print(func(x))
print(func(x+h))
```

```python
x = 0.665
h = 0.001
print(func(x))
print(func(x+h))
```
